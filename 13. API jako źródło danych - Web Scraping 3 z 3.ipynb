```python
import os
import json
import time
import requests
import datetime
import dateutil
import pandas as pd
from dateutil.relativedelta import relativedelta
```

```python
base_url = 'https://api.nytimes.com/svc/archive/v1/'
url = f'{base_url}/2024/3.json?api-key={API_KEY}'

response = requests.get(url).json()
```

```python
response['response']['docs'][:10]
```

```python
response['response']['docs'][0]
```

```python
def is_valid(article):
    has_headline = type(article['headline']) == dict and 'main' in article['headline'].keys()
    return has_headline


def parse(response):
    data = {'headline': [],
            'date': [],
            'doc_type': [],
            'material_type': [],
            'section': [],
            'keywords': []}

    articles = response['response']['docs']
    for article in articles:
        date = dateutil.parser.parse(article['pub_date']).date()
        if is_valid(article):
            data['date'].append(date)
            data['headline'].append(article['headline']['main'])
            if 'section' in article:
                data['section'].append(article['section_name'])
            else:
                data['section'].append(None)
            data['doc_type'].append(article['document_type'])
            if 'type_of_material' in article:
                data['material_type'].append(article['type_of_material'])
            else:
                data['material_type'].append(None)
            keywords = [keyword['value'] for keyword in article['keywords'] if keyword['name'] == 'subject']
            data['keywords'].append(keywords)
    return pd.DataFrame(data)
                                             
```

```python
df = parse(response)
df
```

## Tokenizer

```python
import re

tokenizer = re.compile(r'[\W]+')
tokenized = tokenizer.split(all_headlines)
str(tokenized[:15])
```

```python
tokenized = [word.lower() for word in tokenized]
print(f"Mamy {len(tokenized)} wyraz√≥w")
```

```python
counter = {}

for w in tokenized:
    counter[w] = counter.get(w,0)+1

print(f"Mamy {len(counter.keys())} R√ì≈ªNYCH wyraz√≥w")
```

```python
counted_words = [(word, cnt) for word, cnt in counter.items()]
counted_words[:20]
```

```python
from operator import itemgetter

counted_words.sort(key=itemgetter(1), reverse=True)
counted_words[:20]
```

```python
counted_words[-20:]
```

```python
count_df = pd.DataFrame(counted_words[:120]).reset_index()
columns = list(count_df.columns)
columns[1] = 'word'
columns[2] = 'count'
count_df.columns = columns
count_df
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(20,20))
plt.style.use("dark_background")

chart = sns.lineplot(x='index',
                    y='count',
                    linestyle='-', marker='o',
                    markeredgecolor='black',
                    data=count_df
                   )
chart.set_ylabel('Ilo≈õƒá wystƒÖpie≈Ñ')
chart.set_xlabel('Rank wyrazu')
chart.ticklabel_format(useOffset=True, style='plain')
chart.set_title('Wyrazy w artyku≈Çach NYT')
plt.show()
```
